# How many people have a credit card that expires in 2020

```python
df["CC Exp Date"]
```

```python
def fun():
    count = 0
    for date in df["CC Exp Date"]:
        if date.split("/")[1] == "20":
            count= count + 1
    print(count)
```

```python
fun()
```

# Cheching the same bur using lambda function

```python
len(df[df["CC Exp Date"].apply(lambda x:x[3:] == "20")])
```

# Top 5 most popular email providers

```python
emails = []
for email in df["Email"]:
    emails.append(email.split("@")[1])
```

```python
df["mail_prov"] = emails
```

```python
df.head()
```

```python
df["mail_prov"].value_counts()
```
